## Step 3: Handling Missing Values ( Identify , fill/drop )

```python
df.isnull().sum()
```

```python
df_cleaned = df.copy()
```

```python
df_cleaned['discount'] = df_cleaned['discount'].fillna(0)
df_cleaned['seller_name'] = df_cleaned['seller_name'].fillna("Unknown")
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned['what_customers_said'] = df_cleaned['what_customers_said'].fillna("No review")
df_cleaned['seller_information'] = df_cleaned['seller_information'].fillna("Not available")
```

```python
df_cleaned['initial_price'] = df_cleaned['initial_price'].fillna(0)
df_cleaned['ratings_count'] = df_cleaned['ratings_count'].fillna(0)
```

```python
df_cleaned = df_cleaned.drop(columns=['videos','variations'])
```

```python
df_cleaned.isnull().sum()
```

## Step 4 : Performing Basic Operations ( filter rows , select columns)

```python
df_cleaned[['title', 'final_price', 'rating']].head()
```

```python
df_cleaned[df_cleaned['rating'] > 4]
```

```python
df_cleaned['final_price'] = pd.to_numeric(df_cleaned['final_price'], errors='coerce')
```

```python
df_cleaned[(df_cleaned['rating'] > 4) & (df_cleaned['final_price'] < 100)]
```

```python
df_cleaned.dtypes
```

## Step 5 : Removing Duplicates

```python
df_cleaned = df_cleaned.drop_duplicates()
```

```python
df_cleaned.duplicated().sum()
```

## Step 6 : Feature Engineering ( total_amount = price * quantity )

```python
df_cleaned[['initial_price', 'ratings_count']].info()
```

```python
df_cleaned['initial_price'] = df_cleaned['initial_price'].astype(str)

df_cleaned['initial_price'] = df_cleaned['initial_price'].str.replace('[^0-9.]', '', regex=True)

df_cleaned['initial_price'] = pd.to_numeric(df_cleaned['initial_price'], errors='coerce')
```

```python
df_cleaned['total_amount'] = df_cleaned['initial_price'] * df_cleaned['ratings_count']
```

```python
df_cleaned[['initial_price', 'ratings_count', 'total_amount']].head()
```

## Step 7 : Final Output 

```python
df_final = df_cleaned[['title', 'initial_price', 'ratings_count', 'total_amount']]
```

```python
df_final = df_final.sort_values(by='total_amount', ascending=False)
df_final.head()
```

```python
df_final.to_csv("Cleaned_dataset.csv", index=False)
```
